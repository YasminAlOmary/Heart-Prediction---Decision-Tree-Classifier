# Heart Prediction - Decision Tree Classifier


```python
import pandas as pd # data processing
import numpy as np # working with arrays
import matplotlib.pyplot as plt # visualization

from sklearn.tree import DecisionTreeClassifier as dtc # tree algorithm
from sklearn.model_selection import train_test_split # splitting the data
from sklearn.metrics import accuracy_score # model precision

```

```python
Dataset = pd.read_csv("heart.csv")
```

```python
Dataset.head()
```

```python
Dataset.info()
```

 ### Data Processing

```python
for i in Dataset.Sex.values:
    if i  == 'M':
        Dataset.Sex.replace(i, 0, inplace = True)
    else:
        Dataset.Sex.replace(i, 1, inplace = True)

for i in Dataset.ChestPainType.values:
    if i == 'ATA':
        Dataset.ChestPainType.replace(i, 0, inplace = True)
    elif i == 'NAP':
        Dataset.ChestPainType.replace(i, 1, inplace = True)
    elif i == 'ASY':
        Dataset.ChestPainType.replace(i, 2, inplace = True)
    elif i == 'TA':
        Dataset.ChestPainType.replace(i, 3, inplace = True)
        
for i in Dataset.RestingECG.values:
    if i == 'Normal':
        Dataset.RestingECG.replace(i, 0, inplace = True)
    else:
        Dataset.RestingECG.replace(i, 1, inplace = True)
        
for i in Dataset.ExerciseAngina.values:
    if i == 'N':
        Dataset.ExerciseAngina.replace(i, 0, inplace = True)
    else:
        Dataset.ExerciseAngina.replace(i, 1, inplace = True)
        
for i in Dataset.ST_Slope.values:
    if i == 'Up':
        Dataset.ST_Slope.replace(i, 0, inplace = True)
    else:
        Dataset.ST_Slope.replace(i, 1, inplace = True)        

print(Dataset)
```

```python
Dataset.info()
```

```python
X = Dataset[['Age', 'Sex', 'ChestPainType', 'RestingBP','Cholesterol', 'FastingBS','RestingECG','ExerciseAngina','Oldpeak']]# Features
y = Dataset['HeartDisease']# Target variable

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

```python
# Create Decision Tree classifer object
Dtc = DecisionTreeClassifier()

# Train Decision Tree Classifer
Dtc = Dtc.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = Dtc.predict(X_test)
```

```python
# printing the accuracy of the model
print(accuracy_score(y_pred, y_test))
```
